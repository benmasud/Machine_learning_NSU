```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('mushrooms.csv')
data.head(n=5)
```

#### work flow 
- For each categorical Variable : calculate the frequency ..
- Calculate Accuracy
- Repeating those steps up above for each variable 
- Finding the most informative Variable


```python
target_col = 'class'
categorical_features = data.columns.tolist()  
if target_col in categorical_features:
    categorical_features.remove(target_col)
    
# this line is for them who works hard :D 
#categorical_features = ['cap-shape','cap-surface','cap-color','bruises','odor','gill-attachment','gill-spacing','gill-size','gill-color','stalk-shape','stalk-root','stalk-surface-above-ring','stalk-surface-below-ring','stalk-color-above-ring','stalk-color-below-ring','veil-type','veil-color','ring-number','ring-type','spore-print-color','population','habitat']
```

```python
accuracies = {}
```

```python
for feature in categorical_features:
    frequency_counts = data.groupby([feature, target_col]).size().unstack().fillna(0)

    class_counts = data[target_col].value_counts()
    probabilities = frequency_counts / class_counts
    
    predictions = data[feature].map(probabilities.idxmax())
    accuracy = (predictions == data[target_col]).mean()

    accuracies[feature] = accuracy

most_informative_variable = max(accuracies, key=accuracies.get)

for feature, accuracy in accuracies.items():
    print(f'Variable: {feature}, Accuracy: {accuracy:.4f}')

print(f'Most Informative Variable: {most_informative_variable}')
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(15, 7))
sns.lineplot(x=list(accuracies.keys()), y=list(accuracies.values()), marker="o", color="blue")
plt.xlabel('Categorical Variable')
plt.ylabel('Accuracy')
plt.title('Accuracy of Decision Functions for Categorical Variables')
most_informative_index = list(accuracies.keys()).index(most_informative_variable)
plt.plot(most_informative_variable, accuracies[most_informative_variable], marker="o", color="red", label="Most Informative Variable")
plt.legend()
plt.xticks(rotation=45) 

plt.show()
```

### Foot note

Calculating frequency count 
`.size()`After grouping the data, the size() function is applied to count the number of observations in each group. This counts how many times each combination of feature and target value appears in the dataset. <br>

`fillna(0)` method is used to fill any missing values in the DataFrame with zeros. Missing values might occur if a combination of feature and target variable values doesn't exist in the dataset, and filling with zeros ensures that all possible combinations are accounted for.

`frequency_counts` contains the counts of how many times each unique value of the `feature` corresponds to each class label in the `target_col`
