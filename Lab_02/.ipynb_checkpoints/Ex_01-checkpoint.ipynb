#### Data Manipulation 

```python
import pandas as pd 
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns 

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('mushrooms.csv')
data.head(n=5)
```

```python
data.shape
```

```python
target_col = 'class'
feature_name = 'cap-surface' # replace likewise 

grouped = data.groupby([target_col, feature_name])[feature_name].count()
distribution = grouped / grouped.groupby(level=0).sum()

distribution_df = distribution.unstack()
ax = distribution_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title(f'Distribution of {feature_name} across {target_col}')
plt.xlabel(target_col)
plt.ylabel('Proportion')
plt.legend(title=feature_name, loc='upper right')
plt.show()
```

#### Foot note

`grouped = data.groupby([target_col, feature_name])[feature_name].count()`

`distribution = grouped / grouped.groupby(level=0).sum()` <br>

used to calculate the proportion or distribution of each category of the categorical feature within each class
<br>

`distribution_df = distribution.unstack()` <br>

used to reshape the 'distribution' DataFrame so that it's suitable for creating the bar plot
