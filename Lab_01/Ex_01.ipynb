```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

iris_df= pd.read_csv('iris.csv')
iris_df.head()
```

### Calculate correlations <br>
between features on the entire table and separately by class (use pd.groupby).

```python
columns_to_correlate = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

correlations_dataset = iris_df[columns_to_correlate].corr()
correlations_by_class = iris_df.groupby('Species')[columns_to_correlate].corr()

'''
print("Correlations for the entire dataset:")
print(correlations_matrix)

print("\nCorrelations by class (Species):")
print(correlations_by_class)
'''
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlations_dataset, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Matrix (All species)")
plt.show()
```

```python

for species, correlation_dataset in correlations_by_class.groupby(level=0):
    plt.figure(figsize=(8, 6))
    sns.heatmap(correlation_dataset, annot=True, cmap='coolwarm', linewidths=.5)
    plt.title(f'Correlation Heatmap - {species}')
    plt.show()
```

```python
sns.pairplot(iris_df, hue='Species', palette="Paired")
plt.suptitle("Pair plot of Iris Dataset")
plt.show()
```
