```python
import numpy as np
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris_df= pd.read_csv('iris.csv')
iris_df.head()
```

```python
X = iris_df.iloc[:, 1:5]  
y = iris_df['Species']  
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X, y)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_lda[y == 'Iris-setosa', 0], X_lda[y == 'Iris-setosa', 1], label='Iris Setosa', c='gray', marker='o')
plt.scatter(X_lda[y == 'Iris-versicolor', 0], X_lda[y == 'Iris-versicolor', 1], label='Iris Versicolor', c='orange', marker='s')
plt.scatter(X_lda[y == 'Iris-virginica', 0], X_lda[y == 'Iris-virginica', 1], label='Iris Virginica', c='navy', marker='X')

plt.xlabel('LD1')
plt.ylabel('LD2')
plt.legend(loc='best')
plt.title('Linear Discriminant Analysis (LDA)')
plt.show()
```
